```python
class Car:
    def __init__(self, brand, model, year, mileage):
        self.brand = brand
        self.model = model
        self.year = year
        self.mileage = mileage

    def drive(self, distance):
        self.mileage += distance
        print(f"Car drove {distance} km. New mileage: {self.mileage} km.")

    def display_info(self):
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Year: {self.year}")
        print(f"Mileage: {self.mileage} km")


class ElectricCar(Car):
    def __init__(self, brand, model, year, mileage, battery_capacity, charge_level=100):
        super().__init__(brand, model, year, mileage)
        self.battery_capacity = battery_capacity
        self.charge_level = charge_level

    def charge(self, amount):
        self.charge_level = min(100, self.charge_level + amount)
        print(f"Charged {amount} units. New charge level: {self.charge_level}%.")

    def display_info(self):
        super().display_info()
        print(f"Battery Capacity: {self.battery_capacity} kWh")
        print(f"Charge Level: {self.charge_level}%")


# Example usage:
my_car = Car("Toyota", "Camry", 2022, 10000)
my_electric_car = ElectricCar("Tesla", "Model 3", 2023, 5000, 75)

print("--- Car Info ---")
my_car.display_info()
my_car.drive(200)

print("\n--- Electric Car Info ---")
my_electric_car.display_info()
my_electric_car.drive(150)
my_electric_car.charge(30)
my_electric_car.charge(50) # try to charge over 100%
my_electric_car.display_info()
```

```python
class BankAccount:
    def __init__(self, account_number, balance=0.0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount:.2f}. New balance: ${self.balance:.2f}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ${amount:.2f}. New balance: ${self.balance:.2f}")
        else:
            print("Insufficient funds or invalid withdrawal amount.")

    def get_balance(self):
        return self.balance


class SavingsAccount(BankAccount):
    def __init__(self, account_number, balance=0.0, interest_rate=2.5):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate / 100.0  # Convert percentage to decimal

    def add_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Added interest of ${interest:.2f}. New balance: ${self.balance:.2f}")


# Example usage:
account1 = BankAccount("123456789", 1000.0)
savings_account1 = SavingsAccount("987654321", 5000.0, 3.0)

print("--- Bank Account ---")
account1.deposit(500.0)
account1.withdraw(200.0)
print(f"Account balance: ${account1.get_balance():.2f}")

print("\n--- Savings Account ---")
savings_account1.deposit(1000.0)
savings_account1.withdraw(500.0)
savings_account1.add_interest()
print(f"Savings account balance: ${savings_account1.get_balance():.2f}")

savings_account1.withdraw(10000) #test insufficient funds.
savings_account1.deposit(-100) #test negative deposit.
```
